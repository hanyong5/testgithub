## 타이타닉 데이터의 결측치 처리

```python
titanic_df= pd.read_csv("data/titanic.csv")
titanic_df
```

```python
titanic_df.info()
```

```python
titanic_df.isna().sum()
```

```python
titanic_df['age'] = titanic_df['age'].fillna(titanic_df['age'].mean())
titanic_df['age']
titanic_df
```

```python
titanic_df.info()
```

```python
titanic_df['embark_town'][825:830]
```

```python
# embark_town 열의 최빈값
most_freq = titanic_df['embark_town'].value_counts(dropna=True)
most_freq = titanic_df['embark_town'].value_counts(dropna=True).idxmax()
most_freq
```

```python
(titanic_df['embark_town']=='Southampton').sum()
```

```python
titanic_df.groupby('embark_town').size()
```

```python
titanic_df['embark_town'] = titanic_df['embark_town'].fillna(most_freq)
titanic_df['embark_town'][825:830]
```
