## 7.3 [í”ŒëŸ¬ìŠ¤ ì˜ˆì œ] í”„ë¡œê·¸ë˜ë° ì–¸ì–´ ì„ í˜¸ë„ ì¡°ì‚¬

```python

langs = {"python": 0, "java": 0, "c": 0, "javascript": 0}

survey = ["Python", "Java", "python", "C", "Python"]
for s in survey:
    s = s.lower()
    if s in langs:
        langs[s] += 1

langs
```

### ì—¬ëŸ¬ ì–¸ì–´ ì…ë ¥ ì²˜ë¦¬ (split í™œìš©)

```python

langs = {}
inputs = ["Python, Java", "C, Python", "JavaScript"]

for s in inputs:
    for a in s.lower().split(","):
        a = a.strip()
        langs[a] = langs.get(a, 0) + 1

langs
```
