## 9.2 ë°°ì—´ì˜ í™œìš©
### 9.2-1) xticksë¡œ xì¶• ëˆˆê¸ˆ ìœ„ì¹˜/ì´ë¦„ ì¡°ì ˆ
`np.arange(len(year))`ë¡œ ì¸ë±ìŠ¤ ë°°ì—´ì„ ë§Œë“¤ê³ , xticksì— ëˆˆê¸ˆ ì´ë¦„ì„ yearë¡œ ì„¤ì •í•©ë‹ˆë‹¤. îˆ€fileciteîˆ‚turn6file0îˆ

```python
year = [2006, 2009, 2012, 2015, 2018]
kor =  [547, 546, 554, 524, 526]  # ì˜ˆì‹œ: PISA ìˆ˜í•™(í•œêµ­)

ind = np.arange(len(year))  # [0,1,2,3,4]

plt.plot(ind, kor)
plt.xticks(ind, year)
plt.title("PISA ìˆ˜í•™ ì„±ì (ì˜ˆì‹œ)")
plt.xlabel("Year")
plt.ylabel("Score")
plt.show()
```

### 9.2-2) ì—¬ëŸ¬ ê°œì˜ ë§‰ëŒ€ ê·¸ë˜í”„(ê·¸ë£¹ ë§‰ëŒ€)
ë§‰ëŒ€ ìœ„ì¹˜ë¥¼ `ind-0.2`, `ind+0.2`ì²˜ëŸ¼ ì´ë™í•˜ì—¬ ê²¹ì¹˜ì§€ ì•Šê²Œ í•©ë‹ˆë‹¤. îˆ€fileciteîˆ‚turn6file0îˆ

```python
country = ["Korea", "Japan", "USA", "China"]
male = np.array([174.0, 171.0, 176.5, 172.0])   # ì˜ˆì‹œ
female = np.array([161.0, 158.0, 163.0, 160.0]) # ì˜ˆì‹œ

ind = np.arange(len(country))
w = 0.4

plt.bar(ind - 0.2, male, width=w, label="Male")
plt.bar(ind + 0.2, female, width=w, label="Female")
plt.xticks(ind, country)
plt.title("êµ­ê°€ë³„ í‰ê·  í‚¤(ì˜ˆì‹œ)")
plt.legend()
plt.show()
```

### 9.2-3) ëˆ„ì  ë§‰ëŒ€ ê·¸ë˜í”„(stacked bar)
`bottom=` ì˜µì…˜ìœ¼ë¡œ ì•„ë˜ ë§‰ëŒ€ ìœ„ì— ìŒ“ìŠµë‹ˆë‹¤. îˆ€fileciteîˆ‚turn6file0îˆ

```python
years = ["2017", "2018", "2019", "2020", "2021"]
spring = np.array([250, 260, 240, 230, 255])
summer = np.array([900, 880, 920, 950, 910])
fall   = np.array([320, 300, 310, 280, 290])
winter = np.array([180, 190, 170, 200, 185])

ind = np.arange(len(years))

plt.bar(ind, spring, label="Spring")
plt.bar(ind, summer, bottom=spring, label="Summer")
plt.bar(ind, fall, bottom=spring+summer, label="Fall")
plt.bar(ind, winter, bottom=spring+summer+fall, label="Winter")
plt.xticks(ind, years)
plt.title("ê³„ì ˆë³„ ê°•ìˆ˜ëŸ‰ ëˆ„ì (ì˜ˆì‹œ)")
plt.legend()
plt.show()
```

## 9.2-4) NumPy ìˆ˜í•™ í•¨ìˆ˜
`sin`, `cos` ë“±ì€ ë°°ì—´ ì „ì²´ì— ì ìš©ë©ë‹ˆë‹¤. îˆ€fileciteîˆ‚turn6file0îˆ

```python
x = np.linspace(0, 2*np.pi, 200)
sin_y = np.sin(x)
cos_y = np.cos(x)

plt.plot(x, sin_y, label="sin")
plt.plot(x, cos_y, label="cos")
plt.legend()
plt.title("sin/cos")
plt.show()
```

## 9.2-5) NumPy í†µê³„ í•¨ìˆ˜

### (ì˜ˆì œ) ê³¼ëª©ë³„ í‰ê· ì„ ë§‰ëŒ€ ê·¸ë˜í”„ë¡œ

```python
subjects = ["Math", "English", "Science"]
data = np.array([
    [80, 70, 90],
    [85, 75, 88],
    [90, 65, 92],
    [75, 80, 85],
])  # (í•™ìƒ, ê³¼ëª©)

avg = data.mean(axis=0)
ind = np.arange(len(subjects))

plt.bar(ind, avg)
plt.xticks(ind, subjects)
plt.title("ê³¼ëª©ë³„ í‰ê· (ì˜ˆì‹œ)")
plt.show()
```

## 9.2-7) where í•¨ìˆ˜
ì¡°ê±´ì— ë”°ë¼ ê°’ì„ ì„ íƒí•´ ë°°ì—´ì„ ë°˜í™˜í•©ë‹ˆë‹¤. (ë°°ì—´ ì „ì²´ì— ë¹ ë¥´ê²Œ ì ìš©) îˆ€fileciteîˆ‚turn6file0îˆ

```python
score = np.array([95, 82, 77, 68, 59])
pf = np.where(score >= 70, "pass", "fail")
pf
```

### where ì¤‘ì²© ì˜ˆì œ: A/B/C/D ë“±ê¸‰

```python
score = np.array([95, 82, 77, 68, 59])
grade = np.where(score >= 90, "A",
         np.where(score >= 80, "B",
         np.where(score >= 70, "C", "D")))
grade
```

## LAB 1) 2020ë…„ ì›”ë³„ ì½”ë¡œë‚˜19 ì‹ ê·œ í™•ì§„ì ìˆ˜ ê·¸ë˜í”„(ì˜ˆì‹œ)
ìë£Œì˜ í˜•íƒœì— ë§ì¶° **ì›”ë³„ ì‹ ê·œ í™•ì§„ì**ì™€ **ëˆ„ì  í™•ì§„ì**ë¥¼ ì„  ê·¸ë˜í”„ë¡œ ë‚˜íƒ€ëƒ…ë‹ˆë‹¤. îˆ€fileciteîˆ‚turn6file0îˆ

```python
months = np.arange(1, 13)

# ì˜ˆì‹œ ë°ì´í„°(í•™ìŠµìš©): ì‹¤ì œ ê°’ì´ ì•„ë‹ˆë¼ êµ¬ì¡° ì—°ìŠµìš©ì…ë‹ˆë‹¤.
new_cases = np.array([20, 80, 250, 300, 700, 1200, 1500, 2000, 4000, 6000, 8000, 9000])
cum_cases = np.cumsum(new_cases)

plt.plot(months, new_cases, label="New cases")
plt.plot(months, cum_cases, label="Cumulative")
plt.xticks(months, [f"{m}ì›”" for m in months])
plt.title("2020 ì›”ë³„ ì‹ ê·œ/ëˆ„ì  í™•ì§„ì(ì˜ˆì‹œ)")
plt.legend()
plt.show()
```

## LAB 2) ë™ì „ ë˜ì§€ê¸° ì‹œë®¬ë ˆì´ì…˜
ë˜ì§€ëŠ” íšŸìˆ˜ë¥¼ ëŠ˜ë¦´ìˆ˜ë¡ ì•ë©´ í™•ë¥ ì´ 0.5ì— ê°€ê¹Œì›Œì§€ëŠ” ê²ƒì„ í™•ì¸í•©ë‹ˆë‹¤. îˆ€fileciteîˆ‚turn6file0îˆ

```python
trials = np.arange(1, 5001)  # 1~5000
# 0=ë’·ë©´, 1=ì•ë©´
coin = np.random.randint(0, 2, size=trials[-1])
heads_rate = np.cumsum(coin) / trials

plt.plot(trials, heads_rate)
plt.title("ë™ì „ ë˜ì§€ê¸°: ì•ë©´ í™•ë¥ (ëˆ„ì )")
plt.xlabel("Trials")
plt.ylabel("P(Heads)")
plt.show()
```

## 9.3 [í”ŒëŸ¬ìŠ¤ ì˜ˆì œ] ì¶œìƒì•„ ìˆ˜ ì‹œê°í™”
