## 6.3 [í”ŒëŸ¬ìŠ¤ ì˜ˆì œ] ëª¨ìŠ¤ ë¶€í˜¸ ë²ˆì—­ê¸°

### ì•ŒíŒŒë²³ â†’ ëª¨ìŠ¤ ë¶€í˜¸

```python

alphabet = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
morse = [
    ".-","-...","-.-.","-..",".","..-.","--.","....","..",".---",
    "-.-",".-..","--","-.","---",".--.","--.-",".-.","...","-",
    "..-","...-",".--","-..-","-.--","--.."
]

text = "SOS"
result = ""
for ch in text:
    if ch in alphabet:
        result += morse[alphabet.index(ch)] + " "
print(result)
```

### ëª¨ìŠ¤ ë¶€í˜¸ â†’ ì•ŒíŒŒë²³

```python

m = "... --- ..."
lst = m.split()
result = ""

for code_ in lst:
    if code_ in morse:
        result += alphabet[morse.index(code_)]
print(result)
```
