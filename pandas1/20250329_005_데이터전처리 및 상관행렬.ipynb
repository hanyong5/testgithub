### mpg 데이터셋을 이용

#####  mpg 데이터 읽어오기

```python
mpg_data = sns.load_dataset("mpg")
mpg_data
```

```python
mpg_data.info()
```

```python
mpg_data.isnull().sum()
```

```python
mpg_data['horsepower'].fillna(mpg_data['horsepower'].mean(),inplace=True)
mpg_data.isnull().sum()
```

```python
mpg_data.groupby('origin').size()
```

##### 범주형 데이터(origin)를 숫자로 변환 ('usa':0, 'europe':1, 'japan': 2)

```python
mpg_data['origin_encoded'] = mpg_data['origin'].map({'usa':0, 'europe':1, 'japan': 2})
mpg_data
```

#### 원하는 항목만 선택

```python
subset = mpg_data[['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 
                                                                   'model_year', 'origin_encoded']]
subset
```

##### 표준화

```python
from sklearn.preprocessing import StandardScaler
```

```python
# StandardScaler를 사용하여 데이터 표준화
scaler = StandardScaler()
subset_st = scaler.fit_transform(subset.drop(['mpg'], axis=1)) ## 표준화과정에서 mpg를 빼고해라

subset_st
```

```python
subset_st_df = pd.DataFrame(subset_st,columns=subset.columns[1:])
subset_st_df
```

```python
#### 상관행렬
```

```python
corr_matrix = subset_st_df.corr()
corr_matrix
```

#### 히트맵 그리기

```python
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('MPG 데이터셋의 상관 관계 히트맵')
plt.show()
```

#### tips 데이터셋을 이용

```python
tips = sns.load_dataset("tips")
tips
```

```python
tips.info()
```

```python
num = ["total_bill","tip","size"]
tips[num] = StandardScaler().fit_transform(tips[num])
tips
```

```python
col = ['sex','smoker','day','time']
for r in col:
    tips[r + '_encoded'] =pd.Categorical(tips[r]).codes
##  pd.Categorical : 리스트에 포함된 범주형 변수들을 더미 인코딩으로 변환
tips
```

```python
sns.catplot(x = 'sex', y = 'tip', hue = 'day_encoded', data = tips, kind = 'violin')
```

```python
numeric_cols = tips.select_dtypes(include=['number'])
corr_matrix = numeric_cols.corr()   
corr_matrix
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Tips 데이터의 상관 관계 히트맵 (표준화 및 범주형 데이터 변환 포함)')
plt.show()
```
