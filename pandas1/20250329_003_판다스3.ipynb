### groupby() 와 agg()

```python
# DataFrame.groupby(기준항목).agg(사용하고자 하는 function들)
```

```python
import warnings
warnings.filterwarnings(action='ignore') 

import pandas as pd
```

```python
exam = pd.read_csv('data/exam.csv')
exam
```

```python
exam.groupby("nclass")[['math','english','science']].mean()
```

```python
exam.groupby('nclass').agg(mean_math   = ('math', 'mean'),         
                         sum_math    = ('math', 'sum'),          
                         median_math = ('math', 'median'),        
                         n = ('nclass', 'count')) 
```

```python
exam.query("nclass==1")
```

```python
exam.query("math>50")
```

```python
exam.query("nclass==1 & math>=50")
```

```python
exam.query("nclass == 1 | nclass==3")
```

```python
exam.query("nclass==1")['math'].mean()
```

### 데이터 합치기 _ merge(), concat()

```python
# 중간고사 데이터 만들기
test1 = pd.DataFrame({'id'      : [1, 2, 3, 4, 5, 7],
                      'midterm' : [60, 80, 70, 90, 85, 20]})

# 기말고사 데이터 만들기
test2 = pd.DataFrame({'id'    : [1, 2, 3, 4, 5, 6],
                      'final' : [70, 83, 65, 95, 80, 10]})
```

```python
test1
```

```python
total = pd.merge(test1,test2,on="id",how="left") ##  how='outer'로 하면, 양쪽 모든 데이터
total
```

```python
group_a = pd.DataFrame({'id'   : [1, 2, 3, 4, 5],
                        'test' : [60, 80, 70, 90, 85]})

group_b = pd.DataFrame({'id'   : [6, 7, 8, 9, 10],
                        'test' : [70, 83, 65, 95, 80]})
```

```python
group_all = pd.concat([group_a,group_b],ignore_index=True)
group_all
```

### loc 와  iloc

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'Score': [85, 90, 95]
}
df = pd.DataFrame(data, index=['A', 'B', 'C'])
df
```

```python
df.loc[['A']]
```

```python
df.loc['A':'B','Name']
```

```python
df.loc[:'B','Name']
```

```python
df.loc[df['Age'] > 28] 
```

```python
df.iloc[0:2]
```
